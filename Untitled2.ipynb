```
from math import log, e
def entropy(labels, base=None):
  value,counts = np.unique(labels, return_counts=True)
  norm_counts = counts / counts.sum()
  base = e if base is None else base
  return -(norm_counts * np.log(norm_counts)/np.log(base)).sum()

def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_

# import scipy.stats
# def ent(data):
#     """Calculates entropy of the passed `pd.Series`
#     """
#     p_data = data.value_counts()
#     entropy = scipy.stats.entropy(p_data)
#     return entropy
```

```
transactions_train=pd.read_csv('data/transactions_train.csv')
train_target=pd.read_csv('data/train_target.csv')
info_group = pd.read_csv('data/info_with_big_group.csv', sep=";")
transactions_train = transactions_train.merge(info_group[['small_group_code', 'big_group']], left_on=['small_group'], right_on=['small_group_code'])\
                  .drop(columns=['small_group_code'])
transactions_train['day_week'] = transactions_train['trans_date'] % 7
transactions_train.head()
```

```
agg_features_train=transactions_train.groupby('client_id')['amount_rur'].agg(['sum', 'median', 'mean', 
                                                                              'std', 'min', 'max', 'count', np.ptp,
                                                                               percentile(25), percentile(75), percentile(90)]).reset_index()

agg_train = transactions_train.groupby('client_id')['trans_date']\
                      .agg(['min', 'max', np.ptp, 'nunique', lambda x : np.mean(np.diff(x)), lambda x : np.max(np.diff(x))]).reset_index()
agg_train.rename(columns={'<lambda_0>' : 'mean_diff', '<lambda_1>' : 'max_diff', 'ptp': 'ptp_trans_date'}, inplace=True)
agg_features_train = agg_features_train.merge(agg_train, on='client_id')

agg_features_train['mean_day_price'] = agg_features_train['sum'] / agg_features_train['ptp_trans_date']
agg_features_train['mean_unique_day_price'] = agg_features_train['sum'] / agg_features_train['nunique']

nunique_small_train = transactions_train.groupby('client_id')['small_group'].agg(['nunique']).reset_index().rename(columns={'nunique' : 'nunique_small'})
agg_features_train = agg_features_train.merge(nunique_small_train, on='client_id')
nunique_big_train = transactions_train.groupby('client_id')['big_group'].agg(['nunique']).reset_index().rename(columns={'nunique' : 'nunique_big'})
agg_features_train = agg_features_train.merge(nunique_big_train, on='client_id')
```

```
day_week_rur_train = transactions_train.groupby(['client_id', 'day_week'])['amount_rur'].agg(["sum", "max", "mean", "median", "count"]).reset_index()
day_week_rur_train.fillna(0, inplace=True)
pv_day_week_rur_train=day_week_rur_train.pivot(index='client_id', \
              columns='day_week',values=["sum", "max", "mean", "median", "count"])
pv_day_week_rur_train.fillna(0, inplace=True)
pv_day_week_rur_train.columns=['day_week_rur_'+str(i) for i in pv_day_week_rur_train.columns]
pv_day_week_rur_train = pv_day_week_rur_train.reset_index()
```

```
day_week_small_train = transactions_train.groupby(['client_id', 'day_week'])['small_group'].agg(["nunique"]).reset_index()
day_week_small_train.fillna(0, inplace=True)
pv_day_week_small_train=day_week_small_train.pivot(index='client_id', \
              columns='day_week',values=["nunique"])
pv_day_week_small_train.fillna(0, inplace=True)
pv_day_week_small_train.columns=['day_week_small_'+str(i) for i in pv_day_week_small_train.columns]
pv_day_week_small_train = pv_day_week_small_train.reset_index()
```

```
day_week_big_train = transactions_train.groupby(['client_id', 'day_week'])['big_group'].agg(["nunique"]).reset_index()
day_week_big_train.fillna(0, inplace=True)
pv_day_week_big_train=day_week_big_train.pivot(index='client_id', \
              columns='day_week',values=["nunique"])
pv_day_week_big_train.fillna(0, inplace=True)
pv_day_week_big_train.columns=['day_week_big_'+str(i) for i in pv_day_week_big_train.columns]
pv_day_week_big_train = pv_day_week_big_train.reset_index()
```

```
stat_small_df_train=transactions_train.groupby(['small_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_small_df_train.fillna(0, inplace=True)
stat_small_df_train = stat_small_df_train.reset_index()
stat_small_df_train
```

```
stat_small_df_train=transactions_train.groupby(['small_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_small_df_train.fillna(0, inplace=True)
stat_small_df_train = stat_small_df_train.reset_index()


stat_small_df_train = stat_small_df_train.merge(agg_features_train[['client_id', 'count']], on='client_id', suffixes=('', '_small_procent_all'))
stat_small_df_train['count_small_procent_all'] = stat_small_df_train['count'] / stat_small_df_train['count_small_procent_all']
agg_median_train = transactions_train.groupby(['small_group'])['amount_rur'].agg('mean').reset_index().rename(columns={'amount_rur' : 'mean_small_group'})
stat_small_df_train = stat_small_df_train.merge(agg_median_train, on='small_group')
stat_small_df_train['mean_small_group'] = stat_small_df_train['mean'] / stat_small_df_train['mean_small_group']

agg_sum_small_train = agg_features_train.iloc[:, [0, 1]].rename(columns={'sum': 'sum_small_group'})
stat_small_df_train = stat_small_df_train.merge(agg_sum_small_train, on='client_id')
stat_small_df_train.fillna(0, inplace=True)
stat_small_df_train['sum_small_group'] = stat_small_df_train['sum'] / stat_small_df_train['sum_small_group']

cat_stat_small_train=stat_small_df_train.reset_index().pivot(index='client_id', \
              columns='small_group',values=['count', 'sum', 'min', 'max',  'mean', 
                                            'median', 'std', 'mean_small_group', 
                                            'sum_small_group', 'count_small_procent_all',
                                            'percentile_25', 'percentile_75', 'percentile_90'])
cat_stat_small_train.fillna(0, inplace=True)
cat_stat_small_train.columns=['small_group_'+str(i) for i in cat_stat_small_train.columns]
```

```
stat_big_df_train=transactions_train.groupby(['big_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_big_df_train.fillna(0, inplace=True)
stat_big_df_train = stat_big_df_train.reset_index()


stat_big_df_train = stat_big_df_train.merge(agg_features_train[['client_id', 'count']], on='client_id', suffixes=('', '_big_procent_all'))
stat_big_df_train['count_big_procent_all'] = stat_big_df_train['count'] / stat_big_df_train['count_big_procent_all']
agg_median_train = transactions_train.groupby(['big_group'])['amount_rur'].agg('mean').reset_index().rename(columns={'amount_rur' : 'mean_big_group'})
stat_big_df_train = stat_big_df_train.merge(agg_median_train, on='big_group')
stat_big_df_train['mean_big_group'] = stat_big_df_train['mean'] / stat_big_df_train['mean_big_group']

agg_sum_big_train = agg_features_train.iloc[:, [0, 1]].rename(columns={'sum': 'sum_big_group'})
stat_big_df_train = stat_big_df_train.merge(agg_sum_big_train, on='client_id')
stat_big_df_train.fillna(0, inplace=True)
stat_big_df_train['sum_big_group'] = stat_big_df_train['sum'] / stat_big_df_train['sum_big_group']

cat_stat_big_train=stat_big_df_train.reset_index().pivot(index='client_id', \
              columns='big_group',values=['count', 'sum', 'min', 'max',  'mean', 
                                            'median', 'std', 'mean_big_group', 
                                            'sum_big_group', 'count_big_procent_all',
                                            'percentile_25', 'percentile_75', 'percentile_90'])
cat_stat_big_train.fillna(0, inplace=True)
cat_stat_big_train.columns=['big_group_'+str(i) for i in cat_stat_big_train.columns]
```

```
agg_max_cnt_small_train = transactions_train.groupby(['client_id'])['small_group'].agg(lambda x:x.value_counts().index[0]).to_frame().reset_index()
agg_max_cnt_small_train = agg_max_cnt_small_train.rename(columns={'small_group': 'max_cnt_small_group'})
agg_max_sum_small_train = transactions_train.groupby(['client_id', 'small_group'])['amount_rur'].agg({'sum':'sum'}).reset_index()\
                        .groupby('client_id')['small_group', 'sum'].agg(lambda x: x['small_group'][x['sum'].idxmax()]).reset_index()
agg_max_sum_small_train = agg_max_sum_small_train.drop(columns=['small_group']).rename(columns={'sum': 'max_sum_small_group'})

agg_max_cnt_big_train = transactions_train.groupby(['client_id'])['big_group'].agg(lambda x:x.value_counts().index[0]).to_frame().reset_index()
agg_max_cnt_big_train = agg_max_cnt_big_train.rename(columns={'big_group': 'max_cnt_big_group'})
agg_max_sum_big_train = transactions_train.groupby(['client_id', 'big_group'])['amount_rur'].agg({'sum':'sum'}).reset_index()\
                        .groupby('client_id')['big_group', 'sum'].agg(lambda x: x['big_group'][x['sum'].idxmax()]).reset_index()
agg_max_sum_big_train = agg_max_sum_big_train.drop(columns=['big_group']).rename(columns={'sum': 'max_sum_big_group'})
```

```
entropy_small_group_train = transactions_train.groupby(['client_id'])['small_group'].agg(entropy).to_frame().rename(columns={'small_group': 'entropy_small_group'})
entropy_big_group_train = transactions_train.groupby(['client_id'])['big_group'].agg(entropy).to_frame().rename(columns={'big_group': 'entropy_big_group'})
```

```
train=pd.merge(train_target,agg_features_train, on='client_id')
train=pd.merge(train, agg_max_cnt_small_train, on='client_id')
train=pd.merge(train, agg_max_sum_small_train, on='client_id')
train=pd.merge(train, agg_max_cnt_big_train, on='client_id')
train=pd.merge(train, agg_max_sum_big_train, on='client_id')
train=pd.merge(train, pv_day_week_rur_train, on='client_id')
# train=pd.merge(train, pv_day_week_small_train, on='client_id')
# train=pd.merge(train, pv_day_week_big_train, on='client_id')
train=pd.merge(train, cat_stat_small_train.reset_index(), on='client_id')
train=pd.merge(train, cat_stat_big_train.reset_index(), on='client_id')
train=pd.merge(train, entropy_small_group_train, on='client_id')
train=pd.merge(train, entropy_big_group_train, on='client_id')
train
```

```
%%time
agg_features_test=transactions_test.groupby('client_id')['amount_rur'].agg(['sum', 'median', 'mean', 
                                                                              'std', 'min', 'max', 'count', np.ptp,
                                                                               percentile(25), percentile(75), percentile(90)]).reset_index()

agg_test = transactions_test.groupby('client_id')['trans_date']\
                      .agg(['min', 'max', np.ptp, 'nunique', lambda x : np.mean(np.diff(x)), lambda x : np.max(np.diff(x))]).reset_index()
agg_test.rename(columns={'<lambda_0>' : 'mean_diff', '<lambda_1>' : 'max_diff', 'ptp': 'ptp_trans_date'}, inplace=True)
agg_features_test = agg_features_test.merge(agg_test, on='client_id')

agg_features_test['mean_day_price'] = agg_features_test['sum'] / agg_features_test['ptp_trans_date']
agg_features_test['mean_unique_day_price'] = agg_features_test['sum'] / agg_features_test['nunique']

nunique_small_test = transactions_test.groupby('client_id')['small_group'].agg(['nunique']).reset_index().rename(columns={'nunique' : 'nunique_small'})
agg_features_test = agg_features_test.merge(nunique_small_test, on='client_id')
nunique_big_test = transactions_test.groupby('client_id')['big_group'].agg(['nunique']).reset_index().rename(columns={'nunique' : 'nunique_big'})
agg_features_test = agg_features_test.merge(nunique_big_test, on='client_id')


day_week_rur_test = transactions_test.groupby(['client_id', 'day_week'])['amount_rur'].agg(["sum", "max", "mean", "median", "count"]).reset_index()
day_week_rur_test.fillna(0, inplace=True)
pv_day_week_rur_test=day_week_rur_test.pivot(index='client_id', \
              columns='day_week',values=["sum", "max", "mean", "median", "count"])
pv_day_week_rur_test.fillna(0, inplace=True)
pv_day_week_rur_test.columns=['day_week_rur_'+str(i) for i in pv_day_week_rur_test.columns]
pv_day_week_rur_test = pv_day_week_rur_test.reset_index()

day_week_small_test = transactions_test.groupby(['client_id', 'day_week'])['small_group'].agg(["nunique"]).reset_index()
day_week_small_test.fillna(0, inplace=True)
pv_day_week_small_test=day_week_small_test.pivot(index='client_id', \
              columns='day_week',values=["nunique"])
pv_day_week_small_test.fillna(0, inplace=True)
pv_day_week_small_test.columns=['day_week_small_'+str(i) for i in pv_day_week_small_test.columns]
pv_day_week_small_test = pv_day_week_small_test.reset_index()

day_week_big_test = transactions_test.groupby(['client_id', 'day_week'])['big_group'].agg(["nunique"]).reset_index()
day_week_big_test.fillna(0, inplace=True)
pv_day_week_big_test=day_week_big_test.pivot(index='client_id', \
              columns='day_week',values=["nunique"])
pv_day_week_big_test.fillna(0, inplace=True)
pv_day_week_big_test.columns=['day_week_big_'+str(i) for i in pv_day_week_big_test.columns]
pv_day_week_big_test = pv_day_week_big_test.reset_index()

stat_small_df_test=transactions_test.groupby(['small_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_small_df_test.fillna(0, inplace=True)
stat_small_df_test = stat_small_df_test.reset_index()
stat_small_df_test

stat_small_df_test=transactions_test.groupby(['small_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_small_df_test.fillna(0, inplace=True)
stat_small_df_test = stat_small_df_test.reset_index()


stat_small_df_test = stat_small_df_test.merge(agg_features_test[['client_id', 'count']], on='client_id', suffixes=('', '_small_procent_all'))
stat_small_df_test['count_small_procent_all'] = stat_small_df_test['count'] / stat_small_df_test['count_small_procent_all']
agg_median_test = transactions_test.groupby(['small_group'])['amount_rur'].agg('mean').reset_index().rename(columns={'amount_rur' : 'mean_small_group'})
stat_small_df_test = stat_small_df_test.merge(agg_median_test, on='small_group')
stat_small_df_test['mean_small_group'] = stat_small_df_test['mean'] / stat_small_df_test['mean_small_group']

agg_sum_small_test = agg_features_test.iloc[:, [0, 1]].rename(columns={'sum': 'sum_small_group'})
stat_small_df_test = stat_small_df_test.merge(agg_sum_small_test, on='client_id')
stat_small_df_test.fillna(0, inplace=True)
stat_small_df_test['sum_small_group'] = stat_small_df_test['sum'] / stat_small_df_test['sum_small_group']

cat_stat_small_test=stat_small_df_test.reset_index().pivot(index='client_id', \
              columns='small_group',values=['count', 'sum', 'min', 'max',  'mean', 
                                            'median', 'std', 'mean_small_group', 
                                            'sum_small_group', 'count_small_procent_all',
                                            'percentile_25', 'percentile_75', 'percentile_90'])
cat_stat_small_test.fillna(0, inplace=True)
cat_stat_small_test.columns=['small_group_'+str(i) for i in cat_stat_small_test.columns]


stat_big_df_test=transactions_test.groupby(['big_group', 'client_id'])['amount_rur']\
                        .agg(['count', 'sum', 'min', 'max',  'mean', 'median', 'std', percentile(25), percentile(75), percentile(90)])
stat_big_df_test.fillna(0, inplace=True)
stat_big_df_test = stat_big_df_test.reset_index()


stat_big_df_test = stat_big_df_test.merge(agg_features_test[['client_id', 'count']], on='client_id', suffixes=('', '_big_procent_all'))
stat_big_df_test['count_big_procent_all'] = stat_big_df_test['count'] / stat_big_df_test['count_big_procent_all']
agg_median_test = transactions_test.groupby(['big_group'])['amount_rur'].agg('mean').reset_index().rename(columns={'amount_rur' : 'mean_big_group'})
stat_big_df_test = stat_big_df_test.merge(agg_median_test, on='big_group')
stat_big_df_test['mean_big_group'] = stat_big_df_test['mean'] / stat_big_df_test['mean_big_group']

agg_sum_big_test = agg_features_test.iloc[:, [0, 1]].rename(columns={'sum': 'sum_big_group'})
stat_big_df_test = stat_big_df_test.merge(agg_sum_big_test, on='client_id')
stat_big_df_test.fillna(0, inplace=True)
stat_big_df_test['sum_big_group'] = stat_big_df_test['sum'] / stat_big_df_test['sum_big_group']

cat_stat_big_test=stat_big_df_test.reset_index().pivot(index='client_id', \
              columns='big_group',values=['count', 'sum', 'min', 'max',  'mean', 
                                            'median', 'std', 'mean_big_group', 
                                            'sum_big_group', 'count_big_procent_all',
                                            'percentile_25', 'percentile_75', 'percentile_90'])
cat_stat_big_test.fillna(0, inplace=True)
cat_stat_big_test.columns=['big_group_'+str(i) for i in cat_stat_big_test.columns]


agg_max_cnt_small_test = transactions_test.groupby(['client_id'])['small_group'].agg(lambda x:x.value_counts().index[0]).to_frame().reset_index()
agg_max_cnt_small_test = agg_max_cnt_small_test.rename(columns={'small_group': 'max_cnt_small_group'})
agg_max_sum_small_test = transactions_test.groupby(['client_id', 'small_group'])['amount_rur'].agg({'sum':'sum'}).reset_index()\
                        .groupby('client_id')['small_group', 'sum'].agg(lambda x: x['small_group'][x['sum'].idxmax()]).reset_index()
agg_max_sum_small_test = agg_max_sum_small_test.drop(columns=['small_group']).rename(columns={'sum': 'max_sum_small_group'})

agg_max_cnt_big_test = transactions_test.groupby(['client_id'])['big_group'].agg(lambda x:x.value_counts().index[0]).to_frame().reset_index()
agg_max_cnt_big_test = agg_max_cnt_big_test.rename(columns={'big_group': 'max_cnt_big_group'})
agg_max_sum_big_test = transactions_test.groupby(['client_id', 'big_group'])['amount_rur'].agg({'sum':'sum'}).reset_index()\
                        .groupby('client_id')['big_group', 'sum'].agg(lambda x: x['big_group'][x['sum'].idxmax()]).reset_index()
agg_max_sum_big_test = agg_max_sum_big_test.drop(columns=['big_group']).rename(columns={'sum': 'max_sum_big_group'})

entropy_small_group_test = transactions_test.groupby(['client_id'])['small_group'].agg(entropy).to_frame().rename(columns={'small_group': 'entropy_small_group'})
entropy_big_group_test = transactions_test.groupby(['client_id'])['big_group'].agg(entropy).to_frame().rename(columns={'big_group': 'entropy_big_group'})
```
